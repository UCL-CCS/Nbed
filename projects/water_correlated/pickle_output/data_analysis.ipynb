```python
import pickle
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
cwd = os.getcwd()
```

All the data is stored in a pickled python dictionary with the following keys.

- `bond_length` : list of OH bond lengths
- `global DFT` : list of floats of full system DFT calculations
- `global FCI` : list of floats of full system FCI calculations
- `huz emb FCI` : list of floats of FCI-in-DFT Huzinga PBE calculations
- `mu emb FCI` : list of floats of FCI-in-DFT $\mu$ PBE calculations
- `global H` : list of ints giving number of terms in FCI Hamiltonian at differnt OH bond lengths
- `huz H emb` : list of ints giving number of terms in FCI-in-DFT Huzinga PBE Hamiltonian at differnt OH bond lengths
- `mu H emb` : list of ints giving number of terms in FCI-in-DFT $\mu$  PBE Hamiltonian at differnt OH bond lengths
- `global n qubits` : list of ints giving number of qubits in FCI Hamiltonian at differnt OH bond lengths
- `huz n qubits` :  list of ints giving number of qubits in FCI-in-DFT Huzinga PBE Hamiltonian at differnt OH bond lengths 
- `mu n qubits` :list of ints giving number of qubits in FCI-in-DFT $\mu$ PBE Hamiltonian at differnt OH bond lengths 
- `N active MO` : list of ints giving number of active MOs considered in each embedded calculation
- `global_H` : list of dictionaries of full system FCI Hamiltonian at different OH bond lengths
    - this represents second quantized molecular H as a linear combination of Pauli operators
    - note each key gives Pauli operator and value gives associated coefficient
- `huz_embed_H` : list of dictionaries of FCI-in-DFT Huzinga PBE Hamiltonians at different OH bond lengths
    - this represents second quantized molecular H as a linear combination of Pauli operators
    - note each key gives Pauli operator and value gives associated coefficient
- `mu_embed_HO` : list of dictionaries of FCI-in-DFT $\mu$ PBE Hamiltonians at different OH bond lengths
    - this represents second quantized molecular H as a linear combination of Pauli operators
    - note each key gives Pauli operator and value gives associated coefficient


Note all values are in reference to the different bond lengths in the OH bond length list.

The following code below extracts all the data to plot the figure in the paper.

# SPADE localization

```python
fname = 'active_stretch'
fpath = os.path.join(cwd, fname+'.pickle')
with open(fpath, 'rb') as fp:
    active_stretch = pickle.load(fp)
```

```python
# keys of output data
list(active_stretch.keys())
```

```python
active_st_table = {
    'OH-length': active_stretch['bond_length'],
    'global FCI': active_stretch['global FCI'], 
    'global DFT': active_stretch['global DFT'], 
    
    'huz emb FCI': active_stretch['huz emb FCI'], 
    'mu emb FCI':  active_stretch['mu emb FCI'], 
    
    'global H':  active_stretch['global H'],
    'huz H emb':  active_stretch['huz H emb'], 
    'mu H emb':  active_stretch['mu H emb'], 
    
    'global n qubits':  active_stretch['global n qubits'], 
    'huz n qubits':  active_stretch['huz n qubits'], 
    'mu n qubits': active_stretch['mu n qubits'], 
    'N active MOs':  active_stretch['N active MOs']
                                     }

active_stretch_table = pd.DataFrame(data=active_st_table)
active_stretch_table
```

```python
fname2 = 'enviro_stretch'
fpath2 = os.path.join(cwd, fname2+'.pickle')
with open(fpath2, 'rb') as fp:
    enviro_stretch = pickle.load(fp)
```

```python
enviro_stretch.keys()
```

```python
enviro_st_table = {
    
    'OH-length': enviro_stretch['bond_length'],
    'global FCI': enviro_stretch['global FCI'], 
    'global DFT': enviro_stretch['global DFT'], 
    
    'huz emb FCI': enviro_stretch['huz emb FCI'], 
    'mu emb FCI':  enviro_stretch['mu emb FCI'], 
    
    'global H':  enviro_stretch['global H'],
    'huz H emb':  enviro_stretch['huz H emb'], 
    'mu H emb':  enviro_stretch['mu H emb'], 
    
    'global n qubits':  enviro_stretch['global n qubits'], 
    'huz n qubits':  enviro_stretch['huz n qubits'], 
    'mu n qubits': enviro_stretch['mu n qubits'], 
    'N active MOs':  enviro_stretch['N active MOs']
                                     }

enviro_stretch_table = pd.DataFrame(data=enviro_st_table)
enviro_stretch_table
```

```python
R_list = active_stretch['bond_length']

huz_act_stretch_qubit_H_length_list = [len(H) for H in active_stretch['huz_embed_H']]
mu_act_stretch_qubit_H_length_list  = [len(H) for H in active_stretch['mu_embed_H']]

huz_env_stretch_qubit_H_length_list = [len(H) for H in enviro_stretch['huz_embed_H']]
mu_env_stretch_qubit_H_length_list  = [len(H) for H in enviro_stretch['mu_embed_H']]

global_qubit_H_length_list  = [len(H) for H in active_stretch['global_H']]
```

```python
fig = plt.figure(dpi=300)

gs = fig.add_gridspec(3,2)

ax1 = fig.add_subplot(gs[0,:])# goes over both columns!
ax2 = fig.add_subplot(gs[1,:])
# ax3 = fig.add_subplot(gs[1,1])

# colour_stone = '#C0BDB0' #'#D6D2C4'
# colour_dark_brown = '#4E3629'

# colour_stone = '#500778' 
# colour_dark_brown = '#AC145A'

colour_stone = '#8C8279' 
colour_dark_brown = 'k'#'#47423e'

colour_midblue = '#002855'
colour_brightblue = '#0097A9'

colour_orange = '#EA7600'
colour_dark_yellow = '#F6BE00'

FCI_colour = '#002855'


ax1.plot(R_list,np.log10(abs(np.array(active_st_table['huz emb FCI']) - np.array(active_st_table['global FCI']))), label='Huz emb - act stretch', marker='o',
         color=colour_stone, linestyle='dashed')
ax1.plot(R_list,np.log10(abs(np.array(enviro_st_table['huz emb FCI']) - np.array(enviro_stretch['global FCI']))), label='Huz emb - env stretch', marker='s', 
         color=colour_dark_brown, linestyle='dashed')

ax1.plot(R_list,np.log10(abs(np.array(active_st_table['mu emb FCI']) - np.array(active_st_table['global FCI']))), label='mu emb - act stretch', marker='x', 
         color=colour_orange,linestyle='')
ax1.plot(R_list, np.log10(abs(np.array(enviro_st_table['mu emb FCI']) - np.array(enviro_stretch['global FCI']))), label='Mu emb - env stretch', marker='+', 
         color=colour_dark_yellow, linestyle='')

ax1.plot(R_list,np.log10(abs(np.array(enviro_stretch['global DFT']) - np.array(enviro_stretch['global FCI']))), label='global DFT - env stretch', 
         color=colour_midblue, marker='1')
ax1.plot(R_list,np.log10(abs(np.array(active_st_table['global DFT']) - np.array(active_st_table['global FCI']))), label='global DFT - act stretch', 
         color=colour_brightblue, marker='2', linestyle='dotted')


# ax1.legend(loc="upper right", fontsize =5)

# ax1.set_ylabel('energy error (Ha)')
ax1.set_ylabel(r'$log_{10}(\frac{|\Delta E|}{Ha})$')
# ax1.set_xlabel(r'Bond length ($\AA$)')
ax1.set_xticklabels([])


# First result - number of terms
wid = 0.035

ax2.bar(np.array(R_list)-2*wid, huz_act_stretch_qubit_H_length_list, width=wid, color=colour_stone, align='center', label='$|H_{huz}|$ - act stretch', edgecolor='k', linewidth=0.5)
ax2.bar(np.array(R_list)-wid, mu_act_stretch_qubit_H_length_list, width=wid, color=colour_orange, align='center', label='$|H_{mu}|$ - act stretch', edgecolor='k', linewidth=0.5)
ax2.bar(np.array(R_list), global_qubit_H_length_list, width=wid, color=FCI_colour, align='center', label='$|H_{FCI}|$ - global system', edgecolor='k', linewidth=0.5)
ax2.set_ylabel(r'$|H_{q}|$')
ax2.set_xlabel(r'Bond length ($\AA$)')

ax2.bar(np.array(R_list)+2*wid, huz_env_stretch_qubit_H_length_list, width=wid, color=colour_dark_brown, align='center', label='$|H_{huz}|$- env stretch', edgecolor='k', linewidth=0.5)
ax2.bar(np.array(R_list)+wid, mu_env_stretch_qubit_H_length_list, width=wid, color=colour_dark_yellow, align='center', label='$|H_{mu}|$ - env stretch', edgecolor='k', linewidth=0.5)
ax2.set_xlabel(r'Bond length ($\AA$)')

# shared x axis
ax1.get_shared_x_axes().join(ax1, ax2)

# set axis limits
# ax1.yaxis.set_ticks(np.arange(0, 0.4, 0.1))
ax1.yaxis.set_ticks(np.arange(-2.5, 0, 1))
ax2.yaxis.set_ticks(np.arange(0, 5000, 1000))



plt.tight_layout()


## combine legend
# h1, l1 = ax1.get_legend_handles_labels()
# h2, l2 = ax2.get_legend_handles_labels()
# ax1.legend(h1+h2, l1+l2, bbox_to_anchor=(0.4, -3.0, 0.5, 0.5),  ncol=2)



ax1.legend(bbox_to_anchor=(-0.1, -3.0, 0.5, 0.5), ncol=1)
ax2.legend(bbox_to_anchor=(0.5, -1.25, 0.5, 0.5), ncol=1)


fig.savefig('water_correlated_results_SPADE.tif', bbox_inches = 'tight', transparent=False,
           facecolor='white')
plt.show()
```
